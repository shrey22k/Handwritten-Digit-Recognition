```python
import numpy as np
import matplotlib.pyplot as plt

import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout
```

## Get the data and pre-process it

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
def plot_input_img(i):
    plt.imshow(X_train[i], cmap='binary')
    plt.title(y_train[i])
    plt.axis('off')
    plt.show()
```

```python
for i in range(10):
    plot_input_img(i)
```

```python
## Preprocess the images

## Normalize the images to be between 0 and 1 range
X_train = X_train.astype(np.float32)/255
X_test = X_test.astype(np.float32)/255

## Reshape / expand the dimensions of the images to (28,28,1)
X_train = np.expand_dims(X_train,-1)  # Add channel dimension
X_test = np.expand_dims(X_test,-1)    # Add channel dimension

## Convert the labels to one-hot vectors
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)
```

```python
model = Sequential()
# Add layers to the model
model.add(Conv2D(32, (3,3), input_shape=(28, 28, 1), activation='relu')) # Input layer with 32 filters of size 3x3
model.add(MaxPooling2D((2, 2))) # Downsample the feature maps by 2x2

model.add(Conv2D(64, (3,3), activation='relu'))  # Second convolutional layer with 64 filters
model.add(MaxPooling2D((2, 2))) # Downsample the feature maps by 2x2

model.add(Flatten())  # Flatten the 2D output to 1D
model.add(Dropout(0.25))  # Add dropout for regularization
model.add(Dense(10, activation='softmax'))  # Output layer with softmax activation
```

```python
model.summary()  # Print the model summary
```

```python
model.compile(optimizer='adam', loss=keras.losses.categorical_crossentropy, metrics=['accuracy'])
```

```python
# Callbacks
from keras.callbacks import EarlyStopping, ModelCheckpoint

# Early stopping to prevent overfitting
es = EarlyStopping(monitor='val_accuracy', min_delta=0.01, patience=4, verbose=1, mode='max')

# Model checkpoint to save the best model
mc = ModelCheckpoint('./bestmodel.h5', monitor='val_accuracy', save_best_only=True)

cb = [es, mc] # List of callbacks
```

# Model Training

```python
history = model.fit(X_train, y_train, epochs=50, validation_split=0.3, callbacks=cb)
```
